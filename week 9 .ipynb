optimization in python

if we are having convex optimization problem (constrained or unconstrained) we can solve it using cvxopt library . here only 2 cases of convex programming problem i.e LPP and QPP are covered 

```python
import cvxopt #library for  convex optimization 
import cvxopt.solvers as solvers
```

# solving a linear programming problem 


```python
# a LPP is a convex programming problem
# a caveat : the cvxopt(solvers , matrix) expects the entries of the matrices in floating point numbers
```

```python
print(help(solvers.lp))
```

```python
#format for linear programming problem with only inequality constraints
from cvxopt import matrix,  solvers
A = matrix([[-1.0 , 2.0 , -1.0 , 0.0] , [-1.0 , 3.0 , 0.0 ,-1.0]])
# A  = matrix([[entries in the first column] , [entries in the second column]])
#print(A)
#print(A.size) a is a 4 * 2 matrix 
B = matrix([[-2.0 , 12.0 , 0.0 , 0.0]])  # a 4 * 1 vector 
c = matrix([[2.0 , 3.0]])
sol  = solvers.lp(c , A , B)
print(sol['x'])
# this problem has 4 inequality constraints
```

```python
# format for linear programming problem with inequality & equallity constraints
c = matrix([[1.0  , 3.0]])
G = matrix([[-1.0 , 0.0] ,[0.0 , -1.0]])
h = matrix([[0.0 , 0.0]])
A = matrix([[2.0],[4.0]])
b = matrix([2.0])
sol = solvers.lp(c   , G   ,h , A  , b )
print(sol['x'])
#This problem has 2 inequality constranints and 1 equality constraint
```

```python
#controlling the number of iterations 
from cvxopt import matrix,  solvers
A = matrix([[-1.0 , 2.0 , -1.0 , 0.0] , [-1.0 , 3.0 , 0.0 ,-1.0]])
B = matrix([[-2.0 , 12.0 , 0.0 , 0.0]])
c = matrix([[2.0 , 3.0]])
sol  = solvers.lp(c , A , B , options = {'maxiters':2}) # control the number of iterations
print(sol['x'])
```

# solving a convex quadratic programming problem convex QPP

```python
from cvxopt import matrix , solvers
```

```python
print(help(solvers.qp))
```

```python
P = matrix([[2.0 , 1.0],[1.0 , 4.0]])
print(P)
q = matrix([[3.0 , 4.0]])
g = matrix([[-1.0 , 0.0] , [0.0 , -1.0]])
h = matrix([0.0 , 0.0])
A = matrix([[2.0], [1.0]]) # first column is 2.0 and the second columns is 1.0
b = matrix([1.0])
sol= solvers.qp(P  ,q , g ,h ,A , b)
print(sol['x'])
# a quadratic programming problem with 2 inequality constraints and 1 equality constraint
```
